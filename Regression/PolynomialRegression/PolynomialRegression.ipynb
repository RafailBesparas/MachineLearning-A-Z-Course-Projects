# Problem Statement

The HR department of our company wants to hire a new Staff Software Engineer between levels 6 and 7. He is asking for a salary between 180.000 and 200.000. The first data scientist used Linear regression and to check if the prediction was correct the department asked a colleague to implement another algorithm to check the prediction.

# Dependencies & Feature Label Separation

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import PolynomialFeatures
```

```python
df = pd.read_csv("Position_Salaries.csv")
```

```python
X = df.iloc[:, 1:-1].values
y = df.iloc[:, -1].values
```

# Linear Regression

```python
regressor = LinearRegression()
regressor.fit(X, y)
```

# Polynomial Regression

```python
poly_reg = PolynomialFeatures(degree = 4)
# Matrix where X1 is first feature and X squared is the second feature
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

# Plot Based On Linear Regression

```python
plt.scatter(X, y, color ='red')
# Plot predictions
plt.plot(X, regressor.predict(X), color = 'blue')
plt.title("Truth or False (Linear Regression)")
plt.xlabel("Position Label")
plt.ylabel("Salary")
plt.show()
```

# Plot Based On Polynomial Regression

```python
plt.scatter(X, y, color ='red')
# Plot predictions
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title("Truth or False (Polynomial Regression)")
plt.xlabel("Position Label")
plt.ylabel("Salary")
plt.show()
```

# Visualizing A better curve

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

# Predicting the salary of the Canditate for a posistion between 6 and 7 so it is 6.5

```python
regressor.predict([[6.5]])
```

# Prediction Based On Polynomial Regression

```python
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```

##### So at the end as we could see polynomial regression is clearly more adaptable to our problem and will show a more accurate result than Linear Regression
