# Problem Statement
We try to predict the salaries of the employees based on years of experience. The concept of the project is that with years of experience, the wages of the employee will rapidly grow in any case. The Dataset has one independent variable and one dependent variable.
##### Simple Linear Regression with one feature

# Import Dependencies

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

# Data Preprocessing

```python
df = pd.read_csv("Salary_Data.csv")
```

```python
# Separate the Features from Labels
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

# Test - Train - Split

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 1)
```

```python
regressor = LinearRegression()
```

# Model Training

```python
regressor.fit(X_train, y_train)
```

# Predictions 

```python
y_pred = regressor.predict(X_test)
```

# Model Visualizations

```python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary VS Experience (Training Set)')
plt.xlabel('Years Of Experience')
plt.ylabel('Salary')
plt.show()
```

# Visualizing Test Results

```python
plt.scatter(X_test, y_test, color = 'red')
# We use to judge the results based on the same regression line as the Training results in order to see how 
# close our predictions are
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary VS Experience (Testing Set)')
plt.xlabel('Years Of Experience')
plt.ylabel('Salary')
plt.show()
```
